```python
movie= pd.read_csv("movies.csv")
rating=pd.read_csv("ratings.csv") #ratings is used to make the recomended system
```

```python
import re
```

```python
# making this funtion to clean the title as there are lot of space and brackets 

def clean_title(title):
    title = re.sub("[^a-zA-Z0-9 ]", "", title)
    return title
```

```python
movie["clean_title"]=movie["title"].apply(clean_title)
```

```python
movie
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
# tfTfidfvectorizer turns titles into number
```

```python
vectorizer = TfidfVectorizer(ngram_range=(1,2))      
#instead of looking for just each words ngram look at clusters like toystory
```

```python
#now we turn our vectors into matrices 
movie_matrix = vectorizer.fit_transform(movie["clean_title"])
```

```python
movie_matrix
```

```python
# to find similarity in the terms we are searching for in the movie titles
from sklearn.metrics.pairwise import cosine_similarity
```

```python
#def search(title):
title="Avengers"
title=clean_title(title)
query_vec=vectorizer.transform([title])
similarity=cosine_similarity(query_vec, movie_matrix).flatten()
# Now we find thetitle that have the greatest similarity to the searchec term
indices= np.argpartition(similarity,-10)[-10:] # this will find the 5 most similar term
results= movie.iloc[indices]
```
